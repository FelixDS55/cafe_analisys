```python
import pandas as pd
import sqlite3
import numpy as np
```

```python

con = sqlite3.connect(r'C:\projects\Cafe\cafe.db')
```

```python
df = pd.read_sql("select * from breakfast_at_time", con)
```

```python
df.head()
```

```python
df_amount_ABC = df.groupby('naimenovanie').agg({'amount': sum, 'profit_good': sum}).reset_index()
```

```python
df_amount_ABC['prc_amount'] = round(df_amount_ABC['amount'] / sum(df_amount_ABC['amount']) * 100, 2)
df_amount_ABC['prc_profit'] = round(df_amount_ABC['profit_good'] / sum(df_amount_ABC['profit_good']) * 100, 2)
```

```python
df_amount_ABC = df_amount_ABC.sort_values(by='amount', ascending=False)
```

```python
df_amount_ABC[f'cum_amount_prc'] = df_amount_ABC[f'prc_amount'].cumsum()
```

```python
df_amount_ABC['abc'] = np.where(df_amount_ABC['cum_amount_prc'] < 80, 'A', np.where(df_amount_ABC['cum_amount_prc'] < 95, 'B', 'C'))
```
